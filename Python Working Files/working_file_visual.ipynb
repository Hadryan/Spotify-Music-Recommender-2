```python
import pandas as pd
import pathlib
from ast import literal_eval
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Store filepath in a variable
tracks_path = pathlib.Path("Extraction/tracks.csv")
```

```python
# Read and display the CSV with Pandas
tracks_df = pd.read_csv(tracks_path)
tracks_df
```

```python
# Store filepath in a variable
artists_path = pathlib.Path("Extraction/artists.csv")
```

```python
# Read and display the CSV with Pandas
artists_df = pd.read_csv(artists_path)
artists_df
```

```python
# In tracks_df, convert literal string  in coln "id_artists" to list inside python. Artist ID is a string
tracks_df["artist_ID_list"]=tracks_df['id_artists'].map(literal_eval)
```

```python
# Create Df for relationship table. 2 coln: Artist ID and Genre ID
#Transform each element of a list-like to a row, replicating index values.
relation_df = tracks_df[["id","artist_ID_list"]].explode("artist_ID_list")
relation_df
```

```python
# Merge relational dataframe and artist df based on artist ID

artist_merge_df = pd.merge(relation_df,artists_df,left_on="artist_ID_list", right_on="id")
artist_merge_df
```

```python
# Count results on each Gender ID
artist_merge_df['id_x'].value_counts().sort_values()
```

```python
# In artist_merge_df, convert literal string  in coln "genres" to list inside python. Genres ID is a string

artist_merge_df["genres_list"]=artist_merge_df['genres'].map(literal_eval)
```

```python
tracks_df.loc[tracks_df.id=='5XDh3IwOcwrJjqhfxrapxO']
```

```python
f = lambda x: list(set(z for y in x for z in y))
genres_df = artist_merge_df.groupby(['id_x'])['genres_list'].agg(f).reset_index()
```

```python
# Create dataframe for genres wit unique values

genres_df
```

```python
# Merge genres dataframe and tracks df based on gender ID

track_merge_df = pd.merge(genres_df,tracks_df,left_on="id_x", right_on="id")
track_merge_df
```

```python
# Join Genres ','.join(['a','b','c']) and remove commas
track_merge_df['genres_string']=track_merge_df['genres_list'].map(lambda genres:" ".join(genres))
```
